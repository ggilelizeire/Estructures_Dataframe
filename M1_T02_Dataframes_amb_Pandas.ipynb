## Exercici 1
Descarrega el data set **Airlines Delay: Airline on-time statistics and delay causes** i carrega’l a un pandas Dataframe. Explora les dades que conté, i queda’t únicament amb les columnes que consideris rellevants.

```python
# Almaceno el archivo csv en un dataframe, y lo imprimo:
airlines_df = pandas.read_csv('./input/DelayedFlights.csv')

airlines_df
```

```python
# Elimino la columna "Unnamed 0", correspondiente a los índices del dataframe original, que también los exportaron:
del(airlines_df['Unnamed: 0'])
```

```python
# Compruebo la cantidad de registros y de columnas:
airlines_df.shape
```

```python
# Compruebo el tipo de datos de cada columna:
airlines_df.info()
```

```python
# Compruebo la cantidad de valores null de cada columna:
airlines_df.isna().sum()
```

```python
# Los valores null están en columnas importantes, así que decido eliminar todos los registros con valores null:
airlines_df.dropna(axis=0, how='any', inplace=True)
```

```python
# Compruebo que ya no hay valores null:
airlines_df.isna().sum()
```

```python
# Compruebo que las columnas 'Cancelled', 'Diverted' y 'CancellationCode' ahora tienen un único valor:
print(airlines_df['Cancelled'].unique())
print(airlines_df['Diverted'].unique())
print(airlines_df['CancellationCode'].unique())
```

```python
# Elimino 4 columnas, que tienen un único valor ("Year" desde el principio, y las demás desde que eliminé los nulls):
airlines_df.drop(['Year','Cancelled','CancellationCode','Diverted'], axis=1, inplace=True)
```

```python
# También elimino las columnas 'TaxiIn' y 'TaxiOut', que me parecen poco relevantes:
airlines_df.drop(['TaxiIn','TaxiOut'], axis=1, inplace=True)
```

```python
# Finalmente, actualizo los índices, y vuelvo a imprmir el dataframe:
airlines_df.reset_index(drop=True, inplace=True)

airlines_df
```

## Exercici 2
Fes un informe complet del data set:
- Resumeix estadísticament les columnes d’interès

```python
airlines_df[['ArrDelay','DepDelay','CarrierDelay','WeatherDelay',
             'NASDelay','SecurityDelay','LateAircraftDelay']].describe().round(2)
```

- Troba quantes dades faltants hi ha per columna

```python
# Ya lo he hecho en el ejercicio 1, y he eliminado todos los nulls (ya no hay ninguno).
```

- Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
# Creo la columna 'Velocity', con la velocidad media de cada vuelo:
airlines_df['Velocity'] = (airlines_df['Distance'] / airlines_df['ActualElapsedTime']).round(3)

airlines_df
```

```python
# Creo la columna 'Delayed', iniciándola con valor 'False' en todos los registros:
airlines_df['Delayed'] = False
```

```python
# Itero el dataframe, y actualizo el valor de la columna 'Delayed' por 'True', dependiendo de el caso
#   (si el retraso de llegada es mayor al retraso de salida):

for ind in airlines_df.index:
    if airlines_df.at[ind,'ArrDelay'] > airlines_df.at[ind,'DepDelay']:
        airlines_df.at[ind,'Delayed'] = True

airlines_df
```

- Taula de les aerolínies amb més endarreriments acumulats

```python
airlines_df.groupby('UniqueCarrier')['Delayed'].sum().sort_values(ascending=False)
```

- Quins són els vols més llargs? I els més endarrerits?

```python
# Imprimo el dataframe ordenado por duración del vuelo, de mayor a menor
#   (no he encontrado ningún identificador único de los vuelos, salvo el ID, por eso imprimo todo el dataframe):
airlines_df.sort_values('ActualElapsedTime', ascending=False)
```

```python
# Creo la columna 'TotalDelay', con el retraso de cada vuelo:
airlines_df['TotalDelay'] = airlines_df['ArrDelay'] - airlines_df['DepDelay']

# Imprimo el dataframe ordenado por retraso del vuelo, de mayor a menor
#   (no he encontrado ningún identificador único de los vuelos, salvo el ID, por eso imprimo todo el dataframe):
airlines_df.sort_values('TotalDelay', ascending=False)
```
